```python
# Cell 1: Install & Import tfds
# If you don't have it, run: !pip install tensorflow-datasets
import tensorflow_datasets as tfds
import tensorflow as tf

# Download and load the PlantVillage dataset
# This will automatically download and prepare the data the first time you run it
ds_train, ds_info = tfds.load('plant_village', split='train', with_info=True, as_supervised=True)

# Pre-processing pipeline
def preprocess(image, label):
    # Resize to the input size expected by your CNN (e.g., 150x150)
    image = tf.image.resize(image, (150, 150))
    # Normalize pixel values
    image = tf.cast(image, tf.float32) / 255.0
    return image, label

# Apply pre-processing and batching
ds_train = ds_train.map(preprocess).batch(32).prefetch(tf.data.AUTOTUNE)

print("PlantVillage dataset downloaded and pipeline ready.")
```
